# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров другого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
X_train, X_val, y_train, y_val = train_test_split(data.drop(columns=['Churn']), 
                                                    data['Churn'], test_size=0.33, 
                                                    stratify=data['Churn'], random_state=42)
```

```python
#1. Обработка данных: отнормируйте числовые признаки, а категориальные закодируйте с помощью one-hot-encoding'а.

# Масштабирование численных признаков
scaler = StandardScaler()
X_train_num = scaler.fit_transform(X_train[num_cols])
X_val_num = scaler.transform(X_val[num_cols])

# Кодирование категориальных признаков
cat_enc = OneHotEncoder(drop='first')
X_train_cat = cat_enc.fit_transform(X_train[cat_cols])
X_val_cat = cat_enc.transform(X_val[cat_cols])

# Объединение численных и категориальных признаков
X_train = np.hstack((X_train_cat.toarray(), X_train_num))
X_val = np.hstack((X_val_cat.toarray(), X_val_num))
```

```python
# Обучение линейной модели LogisticRegressionCV.
clf = LogisticRegressionCV(scoring='roc_auc', refit=True, n_jobs=-1).fit(X_train, y_train)
print('roc_auc score %.4f' % clf.score(X_train, y_train))
print('best C %.5f' % clf.C_[0])
```

```python
# Проверим какие результаты даст перебор по  С=100, 10, 1, 0.1, 0.01, 0.001.
clf2 = LogisticRegressionCV(Cs=[100, 10, 1, 0.1, 0.01, 0.001], cv=6, scoring='roc_auc', refit=True, n_jobs=-1).fit(X_train, y_train)
print('roc_auc score %.4f' % clf2.score(X_train, y_train))
print('best C %.5f' % clf2.C_[0])
```

```python
print(roc_auc_score(y_val, clf.predict_proba(X_val)[:, 1]),
      roc_auc_score(y_val, clf2.predict_proba(X_val)[:, 1]))
```

Выпишите какое лучшее качество и с какими параметрами вам удалось получить

```python
print('Лучшее качество roc auc на валидации: %.4f n/ Параметр C = %d' % (clf2.score(X_val, y_val), clf2.C_))
```

```python
X_train, X_val, y_train, y_val = train_test_split(data.drop(columns=['Churn', 'TotalSpent']), 
                                                    data['Churn'], test_size=0.33, 
                                                    stratify=data['Churn'], random_state=42)
```

```python
# Перебор различных Learting rate
```

```python
train_score = []
val_score = []

for l in [0.001, 0.005, 0.007]:
    cbm = CatBoostClassifier(iterations= 5000, learning_rate=l, silent=True, eval_metric='AUC', depth=4)
    cbm.fit(X_train, y_train, cat_features=cat_cols, eval_set=(X_val, y_val))
    train_score.append(cbm.evals_result_['learn']['Logloss'])
    val_score.append(cbm.evals_result_['validation']['Logloss'])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,5))
n = len(cbm.evals_result_['validation']['Logloss'])
l = [0.001, 0.005, 0.007]

for val, train, ax, l in zip(val_score, train_score, axs.flatten(), l):
    ax.plot(np.arange(n), val, marker='.', label='val')
    ax.plot(np.arange(n), train, marker='.', label='train')
    ax.grid(True)
    ax.set_yticks(np.arange(0.2, 0.65, 0.05))
    ax.set_xlabel('iter')
    ax.set_ylabel(f'score. min val score: {np.round(min(val), 4)}')
    ax.set_title(f'learning rate {l}')
    ax.legend(loc='upper right')
    
plt.show()
```

```python
train_score = []
val_score = []

for l in [0.01, 0.03, 0.05]:
    cbm = CatBoostClassifier(iterations= 1000, learning_rate=l, silent=True, eval_metric='AUC', depth=5)
    cbm.fit(X_train, y_train, cat_features=cat_cols, eval_set=(X_val, y_val))
    train_score.append(cbm.evals_result_['learn']['Logloss'])
    val_score.append(cbm.evals_result_['validation']['Logloss'])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15,5))
n = len(cbm.evals_result_['validation']['Logloss'])
l = [0.01, 0.03, 0.05]

for val, train, ax, l in zip(val_score, train_score, axs.flatten(), l):
    ax.plot(np.arange(n), val, marker='.', label='val')
    ax.plot(np.arange(n), train, marker='.', label='train')
    ax.grid(True)
    ax.set_yticks(np.arange(0.2, 0.65, 0.05))
    ax.set_xlabel('iter')
    ax.set_ylabel(f'score. min val score: {np.round(min(val), 4)}')
    ax.set_title(f'learning rate {l}')
    ax.legend(loc='upper right')
    
plt.show()
```
